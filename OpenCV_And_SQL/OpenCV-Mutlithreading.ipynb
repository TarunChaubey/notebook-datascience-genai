```python
import cv2
import uuid
from datetime import datetime
from threading import Thread, 

class StreamVideoPreprocessing:

    @staticmethod
    def ExtractFrame(source_link, cam_id):
        cap = cv2.VideoCapture(source_link)

        while cap.isOpened():
            ret, frame = cap.read()

            frame_id = uuid.uuid4().int
            date_time = datetime.now().strftime("%m-%d-%Y, %H:%M:%S %f")

            frame = cv2.putText(frame, f"{date_time}", (40, 40), cv2.FONT_ITALIC, 0.7, (0, 0, 255), 1)

            data = {"frame_id": frame_id, "cam_id": cam_id, "date_time": date_time}

            print(data, "\n")

            cv2.imshow(f"VideoApps - {cam_id}", frame)

            # Use a small delay (e.g., 1 millisecond)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    source_links = ["http://192.168.1.5:4747/video", "http://192.168.1.2:4747/video"]  # Replace with actual video URLs

    threads = []
    for i, source_link in enumerate(source_links):
        thread = Thread(target=StreamVideoPreprocessing.ExtractFrame, args=(source_link, f'Cam_{i}'))
        threads.append(thread)
        thread.start()

    # Wait for all threads to finish
    for thread in threads:
        thread.join()
```
