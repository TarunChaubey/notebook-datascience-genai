```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('Images/Bird.jpg')
x1,y1 = 180,400
x2, y2 = 1450,1150
img = cv2.rectangle(img,(x1,y1),(x2, y2),(255,0,0),4)
img = cv2.line(img,(x1,y1),(x2, y2),(255,255,0),4)
plt.imshow(img)
```

```python
def AreaOfBox(x1,y1,x2,y2):
    if x1== y2 and x2==y1:
        return (x2-x1) * (y2-y1)
    else:
        return 2*((x2-x1)+(y2-y1))
```

```python
AreaOfBox(x1,y1,x2,y2)
```

```python
img = plt.imread('Images/Bird.jpg')
img = cv2.rectangle(img,(100,300),(1350,1050),(255,0,0),4)
x1,y1 = 180,400
x2, y2 = 1450,1150
img = cv2.rectangle(img,(x1,y1),(x2, y2),(0,0,255),4)
plt.imshow(img)
```
