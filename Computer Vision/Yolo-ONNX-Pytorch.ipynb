```python
from ultralytics import YOLO
```

```python
# Load the YOLO11 model
model = YOLO("yolo11n.pt")
```

```python
model.export(format="openvino")
```

```python
%%time 
results = model("https://ultralytics.com/images/bus.jpg")
```

```python
%%time 
results = model("https://munley.com/wp-content/uploads/People-Crossing-on-Pedestrian-Lane-e1729276338922-1024x457.png.webp")
```

```python
# Export the model to ONNX format
# model.export(format="onnx")  # creates 'yolo11n.onnx'

# Load the exported ONNX model
onnx_model = YOLO("yolo11n.onnx")
```

```python
# Run inference
results = onnx_model("https://ultralytics.com/images/bus.jpg")
```

```python
results = onnx_model("https://munley.com/wp-content/uploads/People-Crossing-on-Pedestrian-Lane-e1729276338922-1024x457.png.webp")
```
